```python
import datetime
```

```python
start = datetime.datetime(2019,3,6)
end = datetime.datetime(2023,3,6)
```

```python
def getdata (tickers):
  data = []
  for ticker in tickers:
    data.append(quandl.get(("BSE/" + ticker),start_date = start , end_date= end).reset_index())
  return data
```
