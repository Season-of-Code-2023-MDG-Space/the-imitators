```python
df = pd.read_csv('/content/drive/MyDrive/MDG SOC 23/NSE Tickers List.csv')
df
```

```python
tickersSensex = df['SYMBOL'].values
tickersSensex
```

```python
# downloading database from yfinance library 
def getdata (tickers):
  data = []
  for ticker in tickers:
    data.append(yf.download(ticker + '.NS').reset_index())
  return data
```

```python
nse = getdata(tickersSensex)
```

```python
# to bring tickers column at first 
for i in range(0,len(nse)):
  cols = nse[i].columns.tolist()
  cols = cols[-1:] + cols[:-1]
  nse[i] = nse[i][cols]

nse[0]
```
